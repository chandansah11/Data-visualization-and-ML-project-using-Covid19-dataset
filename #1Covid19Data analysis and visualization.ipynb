```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\Admin\Desktop\Deep Learning\impl\pmc_1.csv")
```

```python
df
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
df.isnull().sum()
```

```python
x =df["Date"]
y = df["dailyconfirmed"]
plt.plot(x,y)
plt.xticks(rotation = 45)
```

```python

df['Date'] = pd.to_datetime(df['Date'], format='%d.%m.%Y')

plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['dailyconfirmed'], label='Daily Confirmed Cases')
plt.plot(df['Date'], df['dailyrecovered'], label='Daily Recovered Cases')
plt.plot(df['Date'], df['dailydeceased'], label='Daily Deceased Cases')
plt.xlabel('Date')
plt.ylabel('Count')
plt.title('COVID-19 Daily Cases Over Time')
plt.legend()
plt.xticks(rotation=45)
plt.plot()
```

```python
plt.bar(['Total Confirmed', 'Total Recovered', 'Total Deceased'], [df['totalconfirmed'].iloc[-1], df['totalrecovered'].iloc[-1], df['totaldeceased'].iloc[-1]])
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('COVID-19 Total Cases')
plt.show()
```

```python
plt.hist(df['dailyconfirmed'].dropna(), bins=20, color='blue', alpha=0.7)
plt.xlabel('Daily Confirmed Cases')
plt.ylabel('Frequency')
plt.title('Histogram of Daily Confirmed Cases')
plt.show()
```

```python
plt.scatter(df['totalconfirmed'], df['dailyconfirmed'], alpha=0.5)
plt.xlabel('Total Confirmed Cases')
plt.ylabel('Daily Confirmed Cases')
plt.title('Scatter Plot of Total vs. Daily Confirmed Cases')
plt.show()
```

```python
plt.bar(df['Date'], df['dailyconfirmed'], label='Confirmed', color='blue')
plt.bar(df['Date'], df['dailyrecovered'], label='Recovered', color='green', bottom=df['dailyconfirmed'])
plt.bar(df['Date'], df['dailydeceased'], label='Deceased', color='red', bottom=df['dailyconfirmed'] + df['dailyrecovered'])
plt.xlabel('Date')
plt.ylabel('Count')
plt.title('Daily COVID-19 Cases Over Time')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
plt.scatter(df['totalconfirmed'], df['totalrecovered'], c=df['dailyconfirmed'], cmap='viridis', alpha=0.7)
plt.xlabel('Total Confirmed Cases')
plt.ylabel('Total Recovered Cases')
plt.title('Total Confirmed vs. Total Recovered Cases')
plt.xticks(rotation=45)
plt.show()
```

```python
total_dailyconfirmed = df['dailyconfirmed'].sum()
total_dailydeceased = df['dailydeceased'].sum()

mean_dailyconfirmed = df['dailyconfirmed'].mean()
mean_dailydeceased = df['dailydeceased'].mean()

median_dailyconfirmed = df['dailyconfirmed'].median()
median_dailydeceased = df['dailydeceased'].median()

std_dailyconfirmed = df['dailyconfirmed'].std()
std_dailydeceased = df['dailydeceased'].std()

print(f"Total Daily Confirmed Cases: {total_dailyconfirmed}")
print(f"Total Daily Deceased Cases: {total_dailydeceased}")
print(f"Mean Daily Confirmed Cases: {mean_dailyconfirmed}")
print(f"Mean Daily Deceased Cases: {mean_dailydeceased}")
print(f"Median Daily Confirmed Cases: {median_dailyconfirmed}")
print(f"Median Daily Deceased Cases: {median_dailydeceased}")
print(f"Standard Deviation of Daily Confirmed Cases: {std_dailyconfirmed}")
print(f"Standard Deviation of Daily Deceased Cases: {std_dailydeceased}")
```
